## Overall Idea
I will use the basic idea of buy low sell high 

in order to do that I will get the average closing price from the last 100 days, the last 30 days and last day on the market. I will use these 3 data points as markers to see how the stock is doing.

There are 3 possible outcomes:

- If the 100 day average is greater than the 30 day average I assume it will go up and I will buy when it is low
- If the 100 day average is lower than the 30 day average and the last days closing price was up I assume it is too high and the stock will come down so I will sell high
- If the 30 day average is higher than the 100 day average but the last day was below the 30 day average I cant conclude much so I will either hold if i have it or wait another day.

### Testing NKE 

```python
data, meta_data = ts.get_daily(symbol='NKE', outputsize='full')

close_data = data['4. close']

day = close_data[1]


i=1
day5 = 0
while i<5:
    day5 = close_data[i] + day5
    i = i+1
day5 = day5/5
print("Nike 5 day average:", day5)


i=1
month = 0
while i<30:
    month = close_data[i] + month
    i = i+1
month = month/30
print("Nike 30 day average:", month)


i=1
avg = 0
while i<100:
    avg = close_data[i] + avg
    i = i+1
avg = avg/100
print("Nike 100 day average:", avg)


yr = close_data[335] - close_data[365]
print("Nike diff from last year: ", yr)


yr2 = close_data[700] - close_data[730]
print("Nike diff from 2 years: ", yr2)


yr3 = close_data[1065] - close_data[1095]
print("Nike diff from 3 years: ", yr3)


# points
points = 0
if(avg > month):
    points = points + 1
    print (points)
else:
    points = points - 1
    print (points)
    
    
if(day < month):
    points = points + 1
    print (points)
else:
    points = points - 1
    print (points)

    
if(day < day5):
    points = points + 1
    print(points)
else:
    points = points - 1
    print(points)
    
if(day5 < month):
    points = points + 1
    print (points)
else:
    points = points - 1
    print (points)

    
if (yr > 0 and yr2 > 0 and yr3 > 0):
    points = points + 2
    print (points)
elif ((yr > 0 and yr2 > 0) or (yr2 > 0 and yr3 > 0) or (yr > 0 and yr3 > 0)):
    points = points + 1
    print (points)
elif (yr > 0 or yr2 > 0 or yr3 >0):
    print("null")
else:
    points = points - 1
    print (points)
    

#printing
print ("total",points)
if points > 0:
    print("buy NKE")
elif points < 0:
    print("sell NKE")
else:
    print("hold NKE")
```
